```python
import pickle
import matplotlib.pyplot as plt
from skimage.transform import downscale_local_mean, rescale
from skimage.measure import block_reduce
import numpy as np

path = '../data/seg.pickle'
with open(path, 'rb') as f:
    seg = pickle.load(f).astype(int)
# seg = rescale(seg, 0.5, anti_aliasing=False)
# seg = np.rint(seg)

seg = block_reduce(seg, block_size=(8, 8), func=np.max)
```

```python
plt.imshow(seg, cmap='tab20', interpolation='none')
plt.colorbar()
plt.show()
```
